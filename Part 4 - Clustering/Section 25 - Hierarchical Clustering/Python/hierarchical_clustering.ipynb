# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward')) #ward is to connect the points so they have minimum variance, to optimize the results.
plt.title('Dendrogram')
plt.xlabel('No of customers')
plt.ylabel('Euclidean distance')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y = clusters.fit_predict(X)
```

```python
print(y)
```

## Visualising the clusters

```python
plt.scatter(X[y==0, 0], X[y==0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y==1,0], X[y==1,1], s = 100, c = 'yellow', label = 'Cluster 2')
plt.scatter(X[y==2,0], X[y==2,1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y==3,0], X[y==3,1], s = 100, c = 'blue', label = 'Cluster 4')
plt.scatter(X[y==4,0], X[y==4,1], s = 100, c = 'pink' , label = 'CLuster 5')
plt.legend()
plt.show()
```
